## Question 3: Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 
sides of a triangle and returns the area of the triangle. By Heron's formula, the are 
of a triangle with side lengths a, b, and c is s(s − a)(s − b)(s − c), whe e
s = (a + b + c) 2. >>> triangleArea(2, 2,2)

```csharp
using System;

public class Program
{
    public static double TriangleArea(double a, double b, double c)
    {
        double s = (a + b + c) / 2; // Calculate the semi-perimeter
        return Math.Sqrt(s * (s - a) * (s - b) * (s - c)); // Return area using Heron's formula
    }

    public static void Main()
    {
        double area = TriangleArea(2, 2, 2); // Example: all sides equal (equilateral triangle)
        Console.WriteLine("Area of the triangle: " + area);  // Output the area
    }
}

using System;
using System.Collections.Generic;

public class Program
{
    public static void Main()
    {
        // Prompt user for the number of elements
        Console.Write("Input the number of elements to be stored in the array: ");
        int n = Convert.ToInt32(Console.ReadLine());
        
        int[] numbers = new int[n];
        List<int> evenNumbers = new List<int>();
        List<int> oddNumbers = new List<int>();

        // Input elements into the array
        for (int i = 0; i < n; i++)
        {
            Console.Write("Element - " + i + " : "); // Corrected line
            numbers[i] = Convert.ToInt32(Console.ReadLine());
        }

        // Separate even and odd numbers
        foreach (int number in numbers)
        {
            if (number % 2 == 0)
            {
                evenNumbers.Add(number);
            }
            else
            {
                oddNumbers.Add(number);
            }
        }

        // Output results
        Console.WriteLine("The Even elements are:");
        foreach (int even in evenNumbers)
        {
            Console.Write(even + " ");
        }

        Console.WriteLine("\nThe Odd elements are:");
        foreach (int odd in oddNumbers)
        {
            Console.Write(odd + " ");
        }
    }
}

````python
# Question 7 File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic.Write a function bldcount() that reads the file with name name and reports (i.e., prints) how many patients there are in each bloodtype. >>> bldcount('bloodtype.txt')
#There are 10 patients of blood type A.
#There is one patient of blood type B.
#There are 10 patients of blood type AB.
#There are 12 patients of blood type O.
#There are no patients of blood type OO.

def bldcount(filename):
    # Define the blood types to count
    blood_types = ['A', 'B', 'AB', 'O', 'OO']
    
    # Open and read the file
    with open(filename, 'r') as f:
        data = f.read().split()  # Split the file content by spaces
    
    # Count occurrences and print the report
    for blood_type in blood_types:
        count = data.count(blood_type)
        if count == 1:
            print(f"There is {count} patient of blood type {blood_type}.")
        elif count > 1:
            print(f"There are {count} patients of blood type {blood_type}.")
        else:
            print(f"There are no patients of blood type {blood_type}.")

# Call the function with the file name
bldcount('bloodtype1.txt')
````

````python
# Question 8 Write a function curconv() that takes as input: 1. a currency represented using a string (e.g., 'JPY' for the Japanese Yen or 'EUR' for the Euro) 2. an amount
#and then converts and returns the amount in US dollars.
#>>> curconv('EUR', 100)
#122.96544
#>>> curconv('JPY', 100)
#1.241401
def curconv(currency, amount):
    # Open and read the currencies.txt file
    with open('currencies.txt', 'r') as f:
        lines = f.readlines()  # Read all lines from the file
    
    # Find the line with the specified currency and extract the exchange rate
    for line in lines:
        if line.startswith(currency):
            exchange_rate = float(line.split()[1])
            return amount * exchange_rate  # Convert amount to USD
    
    # If currency is not found, return None
    return None

# Testing the function with EUR and JPY
print(curconv('EUR', 100))  # Convert 100 EUR to USD
print(curconv('JPY', 100))  # Convert 100 JPY to USD
````

````python
#Question 10 Encryption is the process of hiding the meaning of a text by substituting letters in the message with other letters, according to some system. If the process is successful, no
#one but the intended recipient can understand the encrypted message. Cryptanalysis
#refers to attempts to undo the encryption, even if some details of the encryption are
#unknown (for example, if an encrypted message has been intercepted). The first step
#of cryptanalysis is often to build up a table of letter frequencies in the encrypted text.
#Assume that the string letters is already defined as
#'abcdefghijklmnopqrstuvwxyz'. Write a function called frequencies()
#that takes a string as its only parameter, and returns a list of integers, showing the
#number of times each character appears in the text. Your function may ignore any
#characters that are not in letters.
#>>> frequencies('The quick red fox got bored and went home.')
#[1, 1, 1, 3, 5, 1, 1, 2, 1, 0, 1, 0, 1, 2, 4, 0, 1, 2, 0, 2,
#1, 0, 1, 1, 0, 0]
#>>> frequencies('apple')
def frequencies(text):
    # Define the string of letters
    letters = 'abcdefghijklmnopqrstuvwxyz'
    
    # Initialize a list to hold the frequency of each letter
    freq = [0] * 26  # 26 letters in the English alphabet
    
    # Convert the text to lowercase to make it case-insensitive
    text = text.lower()
    
    # Count the frequency of each letter in the input text
    for char in text:
        if char in letters:
            index = letters.index(char)  # Find the index of the letter
            freq[index] += 1  # Increment the count for that letter
    
    return freq

# Testing the frequencies function
print(frequencies('The quick red fox got bored and went home.'))
# Expected Output: [1, 1, 1, 3, 5, 1, 1, 2, 1, 0, 1, 0, 1, 2, 4, 0, 1, 2, 0, 2, 1, 0, 1, 1, 0, 0]

print(frequencies('apple'))


````
